```python
import pickle
import pprint

# Đường dẫn tới file
file_path = 'results/multivariate/ETTm2_96/NHITS/hyperopt_eval_train.p'

# Đọc nội dung file
with open(file_path, 'rb') as f:
    trials = pickle.load(f)

# Hiển thị toàn bộ nội dung
print("Số lượng thử nghiệm:", len(trials))
print("Chi tiết toàn bộ các thử nghiệm:")

# Sử dụng pprint để dễ đọc hơn
pprint.pprint(trials.trials)
```

```python
import subprocess

# Chạy lệnh và lấy đầu ra
command = ["python", "-m", "evaluation", "--dataset", "ETTm2", "--horizon", "-1", "--model", "NHITS", "--experiment", "run_1"]
result = subprocess.run(command, capture_output=True, text=True)

# Lưu đầu ra vào một tệp txt
with open("evaluation_results.txt", "w") as file:
    file.write(result.stdout)  # Ghi đầu ra chuẩn vào file
    file.write(result.stderr)  # Ghi thông báo lỗi (nếu có) vào file

print("Kết quả đã được lưu vào evaluation_results.txt")
```
